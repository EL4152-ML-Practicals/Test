```python
import numpy as np
from scipy import stats

data = [90, 94, 53, 79, 94, 53, 65, 87, 90, 70, 69, 65, 89, 85, 53, 47, 61, 27, 80, 53]

mean = np.mean(data)
median = np.median(data)
mode = stats.mode(data).mode
data_range = np.max(data) - np.min(data)

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
print("Range:", data_range)
```
